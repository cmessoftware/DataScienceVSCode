```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Leer los datos
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

test.head()


```

```python
# Seleccionar features simples para empezar
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']
X_train = train[features]
y_train = train['Survived']

print(X_train.head(10))
print(y_train.head(10))
```

```python
# Procesamiento b√°sico: convertir sexo a num√©rico y completar nulos
X_train['Sex'] = X_train['Sex'].map({'male': 0, 'female': 1})
X_train['Age'].fillna(X_train['Age'].median(), inplace=True)
X_train['Fare'].fillna(X_train['Fare'].median(), inplace=True)

X_test = test[features]
X_test['Sex'] = X_test['Sex'].map({'male': 0, 'female': 1})
X_test['Age'].fillna(X_test['Age'].median(), inplace=True)
X_test['Fare'].fillna(X_test['Fare'].median(), inplace=True)

print(X_test.head(10))
```

```python
# Entrenar modelo sencillo
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Predecir para test
predictions = model.predict(X_test)

# Crear archivo submission.csv
submission = pd.DataFrame({
    'PassengerId': test['PassengerId'],
    'Survived': predictions
})
submission.to_csv('submission.csv', index=False)
```
