```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # Fixed import

# Create sample data points
x = np.linspace(0, 10, 100)

# Calculate ni, fi, Ni and Fi curves
ni = np.random.normal(5, 1.5, 100)  # Example normal distribution for frequency
fi = ni/len(ni)  # Relative frequency
Ni = np.cumsum(ni)  # Cumulative frequency 
Fi = Ni/len(ni)  # Cumulative relative frequency

# Create the plot with two y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))  # Added figure size for better visibility
ax2 = ax1.twinx()

# Plot frequencies on left y-axis
ax1.plot(x, ni, 'b-', label='ni (Frequency)')
ax1.plot(x, fi, 'g-', label='fi (Relative Frequency)') 
ax1.set_xlabel('Values')
ax1.set_ylabel('Frequency', color='b')

# Plot cumulative frequencies on right y-axis
ax2.plot(x, Ni, 'r-', label='Ni (Cumulative Frequency)')
ax2.plot(x, Fi, 'y-', label='Fi (Cumulative Relative Frequency)')
ax2.set_ylabel('Cumulative Frequency', color='r')

# Add legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=4)  # Improved legend positioning

plt.title('Frequency Distributions')
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()

plt.hist(fi, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of fi')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

```
