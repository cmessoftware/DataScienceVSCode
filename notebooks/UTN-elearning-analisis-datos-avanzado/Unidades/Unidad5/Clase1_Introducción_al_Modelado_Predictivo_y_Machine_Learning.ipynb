## **1.1 De la Descripción a la Predicción: Jerarquía del Análisis de Datos**

### **Análisis Descriptivo (La Base): ¿Qué pasó?**

* **Ejemplo:** "El mes pasado, vendimos 5,000 unidades del Producto A y el 20% de nuestros clientes premium no renovaron su suscripción."

+ **Herramientas:** Dashboards, reportes, estadísticas básicas (media, mediana).

* **Valor:** Entender el estado actual.

### **Análisis de Diagnóstico: ¿Por qué pasó?**


* **Ejemplo:** "Las ventas del Producto A aumentaron porque lanzamos una campaña en redes sociales que tuvo un 25% más de alcance de lo esperado. Los clientes se fueron porque nuestro competidor principal lanzó una oferta con un 30% de descuento."

* **Herramientas:** Análisis de causa raíz, drill-down en los datos.
+ **Valor:** Identificar las causas de los resultados.

### **Análisis Predictivo: ¿Qué pasará?**

* **Ejemplo:** "Basado en los patrones históricos y las tendencias del mercado, nuestro modelo predice que venderemos 5,500 unidades el próximo mes. Además, el modelo ha identificado a estos 150 clientes como los que tienen la mayor probabilidad de abandonar en los próximos 30 días."
* **Herramientas:** Modelado estadístico, Machine Learning.
* **Valor:** Anticipar eventos futuros y reducir la incertidumbre.

### **Análisis Prescriptivo: ¿Qué deberíamos hacer?**

* **Ejemplo:** "Para prevenir el abandono de esos 150 clientes, el sistema recomienda ofrecerles un descuento personalizado del 15% y un mes de servicio gratuito. Esto maximizará la retención con un costo mínimo."
* **Herramientas:** Optimización, simulación, aprendizaje por refuerzo.
* **Valor:** Recomendar acciones concretas para mejorar los resultados.

## **1.2 ¿Qué es un modelo predictivo?**

Para la toma de decisiones basada en datos, el modelado predictivo se ha convertido en una herramienta valiosa para empresas y organizaciones que buscan **anticipar tendencias, comportamientos y decisiones de mercado futuras**. El modelado predictivo puede utilizar datos históricos, así como datos de encuestas, aplicando algoritmos estadísticos y aprendizaje automático para pronosticar el futuro.

El uso de modelos predictivos abarca una variedad de industrias, desde tecnología, finanzas, gobierno, atención médica hasta marketing y venta minorista, lo que permite a las organizaciones tomar decisiones informadas para aumentar la demanda y el valor de la marca, mejorar la satisfacción del cliente, mejorar el bien público, la eficiencia operativa y lograr una mayor rentabilidad.

El **modelado predictivo y el análisis predictivo** a menudo se utilizan indistintamente, aunque existen diferencias matizadas.

* **Análisis predictivo:**
  - Es un enfoque amplio que utiliza datos históricos y técnicas estadísticas, de aprendizaje automático y minería de datos para predecir resultados futuros y tendencias.
  - Su objetivo es transformar datos en información accionable para la toma de decisiones, identificando riesgos, oportunidades y patrones.
  - Abarca todo el proceso, desde la recopilación de datos hasta la interpretación de los resultados del modelo.
  -Puede utilizar diversos métodos, como modelos de regresión, árboles de decisión, redes neuronales, etc.

* **Modelado Predictivo:**
 - Es una técnica específica dentro del análisis predictivo que se enfoca en la construcción y evaluación de modelos matemáticos que predicen resultados futuros.
 - Implica la selección de variables relevantes, la elección de algoritmos apropiados y la calibración del modelo para que se ajuste a los datos.
 - Se centra en la precisión y la confiabilidad de las predicciones, evaluando el rendimiento del modelo con métricas como el error cuadrático medio o la matriz de confusión.
 - Es un componente esencial del proceso de análisis predictivo, ya que proporciona la herramienta para hacer predicciones basadas en datos



### **1.2.1. Objetivo del modelado predictivo**

El objetivo no es la certeza absoluta, sino cuantificar la probabilidad y reducir la incertidumbre para tomar decisiones más informadas.

## **1.3.1 ¿Qué es el Machine Learning?**

El **Machine Learning o aprendizaje automático** es la rama de la inteligencia artificial que estudia cómo dotar a las máquinas de la capacidad de aprendizaje, entendido este como la generalización del conocimiento a partir de un conjunto de experiencias.

El **Machine Learning** es un componente nuclear que se relaciona y conecta con el resto de categorias. Esto se debe a que el resto de las capacidades pueden ser imitadas, ya sea porque alguien las haya programado o mucho más interesante, porque el propio sistema haya aprendido a realizarla.

## **1.3.2 Cambio de Paradigma**

Lo que hace tan interesante al Machine Learning es que las capacidades de las máquinas pueden ser imitadas de dos maneras: porque alguien las a programado, o, mucho más interesante, porque el propio sistema ha aprendido a realizarlas.

Ejemplos para entender la diferencia:

* Una cosa es programar una máquina para que pueda moverse, y otra muy diferente es programarla para que aprenda a moverse.

* De igual forma, no es lo mismo programar qué elementos conforman una cara, que automáticamente aprender qué es una cara.


**Esta distinción es fundamental y explica por qué es tan común confundir el Machine Learning con la inteligencia artificial en su totalidad**

## **1.3.3 ¿Cómo "Aprende" una Máquina? La Idea de Optimización**

El aprendizaje en la mayoría de los modelos supervisados es un proceso de **optimización**. Imaginen que el modelo tiene una serie de 'perillas' o parámetros que puede ajustar. El objetivo es encontrar la configuración perfecta de esas perillas para que el error de sus predicciones sea lo más bajo posible.

Para hacer esto, necesitamos dos cosas:

1. **Una forma de medir el error:** A esto lo llamamos **Función de Costo o Función de Pérdida (Loss Function)**. Nos dice qué tan 'mal' lo está haciendo el modelo con su configuración actual de perillas.

2. **Una estrategia para ajustar las perillas y reducir ese error:** La estrategia más famosa y fundamental es el **Descenso del Gradiente.**

## **1.3.4 Descenso de Gradiente: Idea intuitiva del algoritmo**

descenso de gradiente.gif

"Imaginen que el error de nuestro modelo es un paisaje montañoso, y queremos encontrar el punto más bajo del valle (el error mínimo). Estamos en una ladera en medio de una densa niebla, por lo que no podemos ver el fondo. ¿Qué hacemos?

El **Descenso del Gradiente** es una estrategia muy intuitiva:
1. **Mira a tu alrededor:** Calcula la pendiente (el gradiente) justo donde estás parado. La pendiente te dice en qué dirección el terreno es más empinado hacia abajo.

2. **Da un pequeño paso:** Da un paso en esa dirección cuesta abajo. El tamaño de este paso se llama **Tasa de Aprendizaje (Learning Rate)**.

3. **Repite:** Desde tu nueva posición, vuelve a mirar a tu alrededor, encuentra la nueva dirección más empinada y da otro paso.

4. **Continúa** hasta que llegues al fondo del valle, donde el terreno es plano (la pendiente es cero).

**En resumen:** El Descenso del Gradiente es un algoritmo iterativo que ajusta los parámetros de un modelo paso a paso, en la dirección que más rápidamente reduce el error. Es el motor que impulsa el entrenamiento de la mayoría de los modelos de Machine Learning, desde la regresión lineal hasta las redes neuronales más complejas."


## **1.3.5 Sobreajuste (Overfitting) y Subajuste (Underfitting):**

overfitting_2.png

Ahora que sabemos cómo aprende un modelo, debemos hablar del principal desafío que enfrentaremos: asegurarnos de que el modelo está aprendiendo patrones generales y no simplemente memorizando los datos de entrenamiento.
Esto se conoce como el Dilema de Sesgo vs. Varianza, que se manifiesta en dos problemas opuestos:"

1. **Underfitting (Subajuste) - Demasiado Simple (Alto Sesgo)**

 * **Analogía:**  Un estudiante que no estudió lo suficiente para el examen. No entiende ni los conceptos básicos.

 * **¿Qué pasa?:** El modelo es demasiado simple para capturar la estructura subyacente de los datos. Falla tanto en los datos de entrenamiento como en los datos nuevos.
 * **Causa:** Usar un modelo muy poco flexible para un problema complejo (ej. usar una línea recta para modelar una curva).

 * **Resultado:** Alto error en entrenamiento y en prueba.

2. **Overfitting (Sobreajuste) - Demasiado Complejo (Alta Varianza)**

  * **Analogía:** Un estudiante que memorizó las respuestas exactas de los ejercicios del libro, pero no entendió los conceptos. Cuando en el examen le ponen un problema ligeramente diferente, no sabe qué hacer.

  * **¿Qué pasa?:** El modelo es tan complejo que se ajusta perfectamente a los datos de entrenamiento, incluyendo el ruido y las peculiaridades aleatorias. Ha memorizado en lugar de generalizar.

  * **Causa:** Usar un modelo demasiado flexible (ej. un polinomio de grado muy alto).

  * **Resultado:** Error muy bajo en entrenamiento, pero error muy alto en datos nuevos (de prueba). Este es uno de los principales problemas al ajustar un modelo de Machine Learning!

3. **Good Fit (Buen Ajuste) - El Equilibrio Perfecto**

  * **Analogía:** Un estudiante que entendió los conceptos y puede aplicarlos a problemas nuevos.

  * **¿Qué pasa?:** El modelo ha capturado la señal verdadera en los datos y ha ignorado el ruido.

  * **Resultado:** Bajo error tanto en los datos de entrenamiento como en los de prueba.

El objetivo de un científico de datos no es minimizar el error en los datos de entrenamiento, sino minimizar el error en datos futuros y no vistos. Por eso, técnicas como la validación cruzada (que veremos en el Tema 5) son cruciales para diagnosticar y prevenir el overfitting.

## **1.3.3. Tipos de Aprendizaje en Machine Learning**

1.  **Aprendizaje Supervisado:** Le damos al modelo datos con "etiquetas" o respuestas correctas.
    *   **Regresión:** Predecir un valor numérico continuo (ej. precio, temperatura).
    *   **Clasificación:** Predecir una categoría (ej. sí/no, spam/no spam, tipo de flor).


regresion_clasificacion.png

2.  **Aprendizaje No Supervisado:** Le damos datos sin etiquetas. El modelo debe encontrar patrones.
    *   **Clustering:** Agrupar datos similares (ej. segmentación de clientes).



Unsupervised-learning.jpg

3. **Aprendizaje por Refuerzo:** El modelo aprende a tomar decisiones en un entorno al recibir recompensas o penalizaciones por sus acciones, optimizando su comportamiento a lo largo del tiempo. (Ej: entrenar un agente para jugar un videojuego)

## **1.3.5 Ciclo de Vida de un Modelo de Machine Learning**

Es un ciclo recurrente ya que los datos se recopilan continuamente, y los modelos se actualizan y mejoran.

**Pasos clave del flujo:**
 1. **Ingestión de datos y control de versiones:** Procesar datos en un formato utilizable y versionar los datos entrantes.

 2. **Validación de datos:** Verificar que las estadísticas de los nuevos datos sean las esperadas y detectar anomalías antes del entrenamiento.

 3. **Preprocesamiento de datos:** Transformar los datos crudos para que puedan ser usados en el modelado (ej: convertir texto en vectores).

 4. **Entrenamiento y ajuste de modelos:** El núcleo del flujo, donde se entrena el modelo para predecir con el menor error posible. El ajuste de modelos, incluyendo hiperparámetros, puede generar mejoras significativas.

 5. **Análisis del modelo:** Evaluación más profunda del rendimiento del modelo, calculando métricas adicionales, verificando la equidad de las predicciones, y la dependencia de las características.

 6. **Control de versiones del modelo:** Rastrear qué modelo, conjunto de hiperparámetros y conjuntos de datos se han seleccionado para la próxima versión.

 7. **Implementación del modelo:** Puesta en producción del modelo entrenado y analizado, preferiblemente usando servidores de modelos modernos para una actualización sencilla.

 8. **Proceso de retroalimentación:** El paso final crucial para medir la eficacia y el rendimiento del modelo implementado y capturar nuevos datos de entrenamiento
