```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset

# sklearn
import sklearn.model_selection as skm
from sklearn.tree import (
    DecisionTreeClassifier as DTC,
    DecisionTreeRegressor as DTR,
    plot_tree,
    export_text
)
from sklearn.metrics import (
    accuracy_score,
    log_loss
)

# ISLP
from ISLP import load_data, confusion_table
from ISLP.models import ModelSpec as MS
```

# Árbol de decisión de clasificiación

#### Importación del dataset y preparación de los datos para el modelo

```python
# se carga el dataset
Carseats = load_data('Carseats')

# a partir de la variable "Sales", generamos el target binario que vamos a predecir
High = np.where(
    Carseats.Sales > 8,
    "Yes",
    "No"
)
```

```python
# se usa MS para preparar el dataset para poder ser consumido por el modelo
model = MS(Carseats.columns.drop('Sales'), intercept=False)
D = model.fit_transform(Carseats)
feature_names = list(D.columns)
X = np.asarray(D)
```

```python
D
```

#### Se entrena el Árbol de decisión

```python
# Se instancia el tipo de árbol
clf = DTC(
    criterion='entropy',
    max_depth=3,
    random_state=0
)

# Se entrena el modelo con las features y el target
clf.fit(X, High)
```

```python
# vemos el accuracy
accuracy_score(High, clf.predict(X))
```

```python
# vemos el desvío
resid_dev = np.sum(log_loss(High, clf.predict_proba(X)))
resid_dev
```

```python
# Graficamos el árbol de decisión

ax = plt.subplots(figsize=(12,12))[1]
plot_tree(
    clf,
    feature_names=feature_names,
    ax=ax
);
```

## Optimización de la poda del Árbol

```python
validation = skm.ShuffleSplit(
    n_splits=1,
    test_size=200,
    random_state=2
)

X_train, X_test, High_train, High_test = skm\
    .train_test_split(
    X,
    High,
    test_size=0.5,
    random_state=0
)

results = skm.cross_validate(
    clf,
    D,
    High,
    cv=validation
)

results
```

```python
# Sacamos el path de poda del árbol
ccp_path = clf.cost_complexity_pruning_path(X_train, High_train)

ccp_alphas = ccp_path.ccp_alphas
impurities = ccp_path.impurities

plt.figure(figsize=(8, 4))
plt.plot(ccp_alphas, impurities, marker='o', drawstyle="steps-post")
plt.xlabel("ccp_alpha")
plt.ylabel("Total Impurity of Leaves")
plt.title("Cost-Complexity Pruning Path")
plt.grid(True)
plt.show()
```

```python
kfold = skm.KFold(
    10,
    random_state=1,
    shuffle=True
)

grid = skm.GridSearchCV(
    clf,
    {'ccp_alpha': ccp_path.ccp_alphas},
    refit=True,
    cv=kfold,
    scoring='accuracy'
)


grid.fit(X_train, High_train)
print(f"best score: {grid.best_score_}")

ax = plt.subplots(figsize=(12, 12))[1]
best_ = grid.best_estimator_
plot_tree(
    best_,
    feature_names=feature_names,
    ax=ax
);
```
