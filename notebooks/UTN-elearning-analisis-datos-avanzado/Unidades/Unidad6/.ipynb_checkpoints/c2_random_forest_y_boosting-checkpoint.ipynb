```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

import sklearn.model_selection as skm
from sklearn.metrics import (
    accuracy_score,
    log_loss
)

# ISLP
from ISLP import load_data, confusion_table
from ISLP.models import ModelSpec as MS
```

```python
# se carga el dataset
Carseats = load_data('Carseats')

# a partir de la variable "Sales", generamos el target binario que vamos a predecir
High = np.where(Carseats.Sales > 8,
    "Yes",
    "No")

# se usa MS para preparar el dataset para poder ser consumido por los modelos
model = MS(Carseats.columns.drop('Sales'), intercept=False)
D = model.fit_transform(Carseats)
feature_names = list(D.columns)
X = np.asarray(D)
```

```python
X_train, X_test, y_train, y_test = skm.train_test_split(
    X, High, test_size=0.3, random_state=42, stratify=High
)
```

```python
bagging_model = RandomForestClassifier(
    n_estimators=10000,
    max_features=1.0,     # usar todos los features → sin submuestreo
    max_depth=None,       # árboles sin podar
    bootstrap=True,       # muestreo con reemplazo
    random_state=42
)
bagging_model.fit(X_train, y_train)
```

```python
rf_model = RandomForestClassifier(
    n_estimators=10000,
    max_features='sqrt',  # submuestreo de features
    max_depth=None,
    bootstrap=True,
    random_state=42
)
rf_model.fit(X_train, y_train)
```

```python
gb_model = GradientBoostingClassifier(
    n_estimators=500,
    learning_rate=0.1,
    max_depth=None,
    random_state=42
)
gb_model.fit(X_train, y_train)
```

```python
models = {
    "Bagging": bagging_model,
    "Random Forest": rf_model,
    "Gradient Boosting": gb_model
}

for name, model in models.items():
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {acc:.3f}")
```

```python
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
plt.title("Importancia de variables (Random Forest)")
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), np.array(feature_names)[indices], rotation=90)
plt.tight_layout()
plt.show()
```

```python
importances = bagging_model.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
plt.title("Importancia de variables (Bagging)")
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), np.array(feature_names)[indices], rotation=90)
plt.tight_layout()
plt.show()
```
