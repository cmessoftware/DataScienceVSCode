## Abordaje para dataset muy desbalanceados como en este caso (0 -> 92,5%, 1 -> 8,5%)
1Ô∏è‚É£ Re-muestreo (Resampling)

Incluye algoritmos para:

Oversampling ‚Üí aumentar las muestras de la clase minoritaria para equilibrar las proporciones.

Ejemplo: RandomOverSampler (duplica aleatoriamente ejemplos minoritarios).

Ejemplo avanzado: SMOTE (Synthetic Minority Oversampling Technique), que crea nuevas muestras sint√©ticas interpolando entre puntos reales minoritarios.

Undersampling ‚Üí reducir la clase mayoritaria eliminando ejemplos.

Ejemplo: RandomUnderSampler.

2Ô∏è‚É£ Algoritmos h√≠bridos

Combinan oversampling + undersampling, o SMOTE + limpieza de ruido (ej. SMOTETomek, SMOTEENN).

3Ô∏è‚É£ Versiones adaptadas de Pipeline

imblearn trae su propia versi√≥n de Pipeline (imblearn.pipeline.Pipeline o ImbPipeline) que permite insertar pasos de re-muestreo en medio del pipeline.
Esto no es posible con el Pipeline est√°ndar de sklearn, porque sklearn asume que el n√∫mero de filas no cambia entre pasos (y un oversampler justamente cambia el n√∫mero de filas).

Ejemplo:

```python
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression

pipe = ImbPipeline(steps=[
    ('smote', SMOTE(random_state=42)),
    ('clf', LogisticRegression())
])
```

Ese pipeline:

Aplica SMOTE para equilibrar las clases en el conjunto de entrenamiento.

Luego entrena el modelo sobre los datos balanceados.

4Ô∏è‚É£ Integraci√≥n directa con sklearn

La API de imblearn est√° hecha para que se use igual que sklearn, as√≠ que pod√©s:

Meter SMOTE o cualquier sampler en un GridSearchCV.

Usar validaci√≥n cruzada con datos re-muestreados solo en los folds de entrenamiento.

Incluir re-muestreo como parte de un pipeline reproducible.

üí° En resumen: imblearn es el ‚Äúkit‚Äù para manejar desbalance de clases dentro de sklearn de forma elegante y reproducible.
En nuestro driver lo usamos porque queremos que el paso de SMOTE est√© dentro del pipeline, de modo que se aplique solo al entrenamiento y quede registrado en MLflow junto con el modelo.


## Evaluaci√≥n con NearMiss
NearMiss es un m√©todo de submuestreo (undersampling) de la clase mayoritaria basado en vecinos m√°s cercanos. En vez de tirar negativos al azar, elige los negativos ‚Äúm√°s cercanos‚Äù a los positivos para equilibrar el set y ‚Äúapretar‚Äù la frontera de decisi√≥n.

¬øC√≥mo funciona?

Tres variantes cl√°sicas (en imblearn.under_sampling.NearMiss):

NearMiss-1: se queda con negativos cuyo promedio de distancia a los k positivos m√°s cercanos es m√≠nimo (cerca de positivos).

NearMiss-2: negativos cuyo promedio de distancia a los k positivos m√°s lejanos es m√≠nimo (cubren el ‚Äúborde externo‚Äù).

NearMiss-3: primero, para cada positivo elige sus m negativos m√°s cercanos; luego submuestrea de ese pool.

Par√°metros √∫tiles: version={1,2,3}, n_neighbors (k para v1/v2) y n_neighbors_ver3 (m para v3).

Pros / Contras (r√°pido)

‚úÖ Focaliza en frontera ‚Üí puede mejorar recall de la minor√≠a.

‚úÖ M√°s informativo que RandomUnderSampler.

‚ö†Ô∏è Puede sobreajustar si el espacio est√° ruidoso.

‚ö†Ô∏è Reduce datos de la mayoritaria ‚Üí riesgo de p√©rdida de informaci√≥n.

‚ö†Ô∏è Es basado en distancias ‚Üí escal√° num√©ricas antes; con OHE de mucha cardinalidad puede volverse ruidoso.

D√≥nde colocarlo en tu flujo

Siempre dentro de un Pipeline y despu√©s del preprocesamiento (imputaci√≥n, OHE, escalado si aplica) y antes del clasificador. As√≠ se ejecuta solo en los folds de train (sin leakage).

````python
## 5.5 # ====================== MAPEO: PREPROCESADOR x MODELO ======================
# Reusa pre_logreg, pre_knn, pre_nb, pre_tree, pre_gb

def get_models_and_pres():
    models = {}

    # --- Group "others" ---
    models["Logistic_Regression"] = {"pre": pre_logreg, "clf": LogisticRegression(max_iter=2000), "group": "others"}
    models["KNeighborsClassifier"] = {"pre": pre_knn, "clf": KNeighborsClassifier(n_neighbors=15), "group": "others"}
    models["SVC"] = {"pre": pre_logreg, "clf": SVC(kernel="rbf", probability=True, random_state=RANDOM_STATE), "group": "others"}
    models["SVC_dense"]  = {"pre":pre_svc_dense, "clf":  SVC(kernel="rbf", probability=True, random_state=RANDOM_STATE), "group": "others"}
    models["SVC_sparse"] = {"pre":pre_svc_sparse, "clf": SVC(kernel="rbf", probability=True, random_state=RANDOM_STATE), "group": "others"}
    models["GaussianNB"] = {"pre": pre_nb, "clf": GaussianNB(), "group": "others"}

    # --- Group "trees" ---√Årboles / ensambles ---
    models["DecisionTreeClassifier"] = {"pre": pre_tree, "clf": DecisionTreeClassifier(random_state=RANDOM_STATE), "group": "trees"}
    models["RandomForestClassifier"] = {"pre": pre_tree, "clf": RandomForestClassifier(n_estimators=400, random_state=RANDOM_STATE, n_jobs=-1), "group": "trees"}
    models["GradientBoostingClassifier"] = {"pre": pre_gb, "clf": GradientBoostingClassifier(random_state=RANDOM_STATE), "group": "trees"}

    # --- Boosters (si est√°n disponibles) ---
    if XGBClassifier is not None:
        models["XGBClassifier"] = {"pre": pre_gb, "clf":XGBClassifier(
            n_estimators=600, learning_rate=0.05, subsample=0.8,
            colsample_bytree=0.8, max_depth=6, eval_metric="logloss",
            n_jobs=-1, tree_method="hist", random_state=RANDOM_STATE
        ), "group": "trees"}
    if LGBMClassifier is not None:
        models["LGBMClassifier"] = {"pre":pre_gb, "clf":LGBMClassifier(
            n_estimators=800, learning_rate=0.05, random_state=RANDOM_STATE, n_jobs=-1
        ), "group": "trees"}
    if CatBoostClassifier is not None:
        models["CatBoostClassifier"] = {"pre":pre_gb, "clf":CatBoostClassifier(
            iterations=800, depth=6, learning_rate=0.05,
            loss_function='Logloss', verbose=False, random_state=RANDOM_STATE
        ), "group": "trees"}

    return models
````

## Nota: 
¬øPor qu√© SVC usa pre_logreg?

SVC (RBF) es muy sensible al scaling de las variables num√©ricas ‚Üí pre_logreg ya hace impute + StandardScaler en num√©ricas (lo esencial).

En pre_logreg las categ√≥ricas quedan como OHE esparso (0/1) y no se escalan.
Para SVC esto suele estar bien: no necesit√°s escalar OHE y evit√°s densificar la matriz (que explota memoria/tiempo).

pre_knn en cambio fuerza denso (OHE sparse_output=False) y escala todo (num/bin/cat). Eso tiene sentido para KNN (distancias puras), pero no aporta tanto a SVC y penaliza rendimiento.

En resumen: con SVC me interesa escalar fuertes a las num√©ricas y mantener OHE esparso; pre_logreg ya cumple eso con buen costo/beneficio.

### Para probar alternativas
A) Mantener esparso pero ‚Äúbalancear‚Äù magnitudes de todos los features

Us√° un post-escale que preserve sparsity:

```python
from sklearn.preprocessing import MaxAbsScaler

pre_svc_sparse = pre_logreg  # mismo pipeline
svc_pipe = Pipeline([
    ("pre", pre_svc_sparse),
    ("post_scale", MaxAbsScaler()),   # escala cada columna a [-1, 1] sin densificar
    ("classifier", SVC(kernel="rbf", probability=True, random_state=RANDOM_STATE)),
])
```

Esto iguala rangos sin centrar (no rompe la sparsity) y a veces mejora estabilidad.

B) Versi√≥n densa tipo KNN (m√°s pesada)

Solo √∫til si sospech√°s que OHE sin escalar te est√° sesgando distancias:

### OHE denso + scaler en cat/bin (cuidado memoria/tiempo)
models["SVC_dense"] = (pre_knn, SVC(kernel="rbf", probability=True, random_state=RANDOM_STATE))

### Pragmatismo
Empezar con pre_logreg + SVC (r√°pido y s√≥lido).

Si se observan mejoras marginales necesarias, probar MaxAbsScaler despu√©s del ColumnTransformer.

Dejar la versi√≥n densa como experimento, no default.


````python
#5.6 ====================== BUILDER DE PIPELINES ======================
def build_pipeline(pre, clf, imbalance_strategy="NONE", use_winsor=False):
    preproc = add_winsor_to_numeric(pre) if use_winsor else pre

    # NONE y CLASS_WEIGHT no usan sampler
    if imbalance_strategy in ("NONE", "CLASS_WEIGHT"):
        return Pipeline([("pre", preproc), ("classifier", clf)])

    # SMOTE / NEARMISS: requieren to_dense + sampler
    steps = [("pre", preproc), ("to_dense", ensure_dense())]
    if imbalance_strategy == "SMOTE":
        from imblearn.over_sampling import SMOTE
        steps.append(("sampler", SMOTE(random_state=RANDOM_STATE, k_neighbors=5)))
    elif imbalance_strategy == "NEARMISS":
        from imblearn.under_sampling import NearMiss
        steps.append(("sampler", NearMiss(version=1, n_neighbors=3)))
    else:
        raise ValueError("imbalance_strategy inv√°lida")

    steps.append(("classifier", clf))
    from imblearn.pipeline import Pipeline as ImbPipeline
    return ImbPipeline(steps)
````

````python
#5.7 ====================== LOOP PRINCIPAL ======================
def run_all_strategies_with_mlflow(X_train, y_train, X_test, y_test,
                                   enable_winsor=False, experiment_name="Modeling Sweep"):
    mlflow.set_experiment(experiment_name)

    pos_rate = float(np.mean(y_train))
    n_pos = int(np.sum(y_train))
    n_neg = int(len(y_train) - n_pos)
    pos_weight = (n_neg / max(n_pos, 1))  # para scale_pos_weight o dict weights

    models = get_models_and_pres()
    leaderboard = []

    with mlflow.start_run(run_name="FULL_SWEEP", nested=False):
        mlflow.set_tags({"pos_rate_train": f"{pos_rate:.5f}", "n_pos": n_pos, "n_neg": n_neg})

        for model_name, cfg in models.items():
            pre, clf, group = cfg["pre"], cfg["clf"], cfg["group"]

            with mlflow.start_run(run_name=f"{model_name}", nested=True):
                for strategy in ["NONE", "SMOTE", "NEARMISS", "CLASS_WEIGHT"]:
                    run_name = f"{model_name} | BAL={strategy} | WINSOR={'ON' if enable_winsor else 'OFF'}"
                    with mlflow.start_run(run_name=run_name, nested=True):
                        clf_to_use = clone(clf)
                        cw_applied, cw_details = False, {}

                        # Solo en CLASS_WEIGHT ajustamos el clasificador
                        if strategy == "CLASS_WEIGHT":
                            clf_to_use, cw_applied, cw_details = apply_class_weight(clf_to_use, pos_weight)

                        tags = {
                            "model_name": model_name,
                            "model_group": group,
                            "imbalance_strategy": strategy,
                            "winsor": str(enable_winsor),
                            "pos_rate_train": f"{pos_rate:.5f}",
                            "class_weight_applied": str(cw_applied),
                        }
                        if cw_details:
                            # ejemplo: "class_weight=balanced" o "scale_pos_weight=12.5"
                            tags["class_weight_details"] = ";".join(f"{k}={v}" for k, v in cw_details.items())

                        mlflow.set_tags(tags)

                        pipe = build_pipeline(pre, clf_to_use,
                                              imbalance_strategy=strategy,
                                              use_winsor=enable_winsor)

                        # params del clasificador
                        try:
                            cls = pipe.named_steps.get("classifier", None)
                            if cls is None and hasattr(pipe, "steps"):
                                for n, est in pipe.steps:
                                    if n == "classifier":
                                        cls = est; break
                            if cls is not None:
                                safe_log_params("classifier__", cls)
                        except Exception:
                            pass

                        # ENTRENAR + M√âTRICAS (usa tu util; si no acepta ese orden, usa el adapter)
                        row = mlu.evaluate_and_log(pipe, X_train, y_train, X_test, y_test, tags)
                        leaderboard.append(row)

        # Resumen √∫nico (incluye CLASS_WEIGHT autom√°ticamente)
        _ = compare_visualizations_balancing(
            pd.DataFrame(leaderboard),
            primary_metric="avg_precision_test",
            artifact_prefix="diabetes_balancing_summary",
            run_name="BALANCING_SUMMARY"
        )

    df_leader = pd.DataFrame(leaderboard).sort_values(
        ["avg_precision_test", "roc_auc_test", "f1_at_opt_thr"], ascending=False
    ).reset_index(drop=True)

    print(df_leader.head(20))
    return df_leader
````

````python
##5.9 ===================================COMPARACION Y VISUALIZACION ====================================

import mlflow
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def compare_visualizations_balancing(
    leaderboard_df: pd.DataFrame,
    primary_metric: str = "avg_precision_test",
    artifact_prefix: str = "balancing_comparison",
    run_name: str = "BALANCING_SUMMARY",
):
    """
    Visualiza y registra en MLflow un resumen comparativo para tu barrido:
    - Espera columnas: ['model_name','imbalance_strategy','roc_auc_test','avg_precision_test',
                        'f1_at_opt_thr','precision_at_opt_thr','recall_at_opt_thr','opt_threshold']
      Opcional: 'winsor' (ON/OFF)
    - primary_metric: m√©trica para decidir el 'best overall' (por defecto AP).
    """

    required = {
        "model_name", "imbalance_strategy", "roc_auc_test", "avg_precision_test",
        "f1_at_opt_thr", "precision_at_opt_thr", "recall_at_opt_thr", "opt_threshold"
    }
    missing = required - set(leaderboard_df.columns)
    if missing:
        raise ValueError(f"Faltan columnas en leaderboard_df: {missing}")

    df = leaderboard_df.copy()

    # Etiqueta legible: Modelo | Estrategia | (W=ON/OFF si existe)
    if "winsor" in df.columns:
        df["winsor_str"] = df["winsor"].astype(str).str.upper()
        df["label"] = df["model_name"] + " | " + df["imbalance_strategy"] + " | W=" + df["winsor_str"]
    else:
        df["label"] = df["model_name"] + " | " + df["imbalance_strategy"]

    # Orden por m√©trica primaria
    if primary_metric not in df.columns:
        raise ValueError(f"primary_metric '{primary_metric}' no est√° en el DataFrame.")
    df_sorted = df.sort_values(primary_metric, ascending=False).reset_index(drop=True)

    # Ganadores
    best_row = df_sorted.iloc[0]
    best_label = str(best_row["label"])
    best_model = str(best_row["model_name"])
    best_strategy = str(best_row["imbalance_strategy"])
    best_primary_value = float(best_row[primary_metric])

    # Agregados por estrategia
    agg = df.groupby("imbalance_strategy").agg(
        mean_AP=("avg_precision_test", "mean"),
        mean_AUC=("roc_auc_test", "mean"),
        mean_F1opt=("f1_at_opt_thr", "mean"),
        mean_Popt=("precision_at_opt_thr", "mean"),
        mean_Ropt=("recall_at_opt_thr", "mean"),
        n=("model_name", "count"),
    ).reset_index()

    # Figura 2x3
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle("Comparaci√≥n de Modelos - Desbalanceo (NONE vs SMOTE vs NearMiss)", fontsize=16, fontweight="bold")

    # 1) AP por label
    top = df_sorted.copy()
    axes[0, 0].barh(top["label"], top["avg_precision_test"])
    axes[0, 0].invert_yaxis()
    axes[0, 0].set_xlabel("Average Precision (AP)")
    axes[0, 0].set_title("AP por Modelo/Estrategia")
    axes[0, 0].grid(axis="x", alpha=0.3)

    # 2) AUC por label
    axes[0, 1].barh(top["label"], top["roc_auc_test"])
    axes[0, 1].invert_yaxis()
    axes[0, 1].set_xlabel("ROC AUC")
    axes[0, 1].set_title("AUC por Modelo/Estrategia")
    axes[0, 1].grid(axis="x", alpha=0.3)

    # 3) F1@opt por label
    axes[0, 2].barh(top["label"], top["f1_at_opt_thr"])
    axes[0, 2].invert_yaxis()
    axes[0, 2].set_xlabel("F1 @ umbral √≥ptimo")
    axes[0, 2].set_title("F1@opt por Modelo/Estrategia")
    axes[0, 2].grid(axis="x", alpha=0.3)

    # 4) Prec@opt vs Rec@opt (tama√±o ~ AP)
    sizes = 300 * (top["avg_precision_test"] - top["avg_precision_test"].min() + 1e-6) / \
            (top["avg_precision_test"].max() - top["avg_precision_test"].min() + 1e-6) + 50
    sc = axes[1, 0].scatter(top["precision_at_opt_thr"], top["recall_at_opt_thr"], s=sizes, alpha=0.7)
    for i, row in top.iterrows():
        axes[1, 0].annotate(row["label"].split("|")[0].strip()[:10],
                            (row["precision_at_opt_thr"], row["recall_at_opt_thr"]),
                            xytext=(5, 5), textcoords="offset points", fontsize=8)
    axes[1, 0].set_xlabel("Precisi√≥n @ opt")
    axes[1, 0].set_ylabel("Recall @ opt")
    axes[1, 0].set_title("Precisi√≥n vs Recall (tama√±o ~ AP)")
    axes[1, 0].grid(True, alpha=0.3)

    # 5) Heatmap de AP: filas=modelo, cols=estrategia
    pivot_ap = df.pivot_table(index="model_name", columns="imbalance_strategy",
                              values="avg_precision_test", aggfunc="mean")
    im = axes[1, 1].imshow(pivot_ap.values, aspect="auto")
    axes[1, 1].set_xticks(range(pivot_ap.shape[1]))
    axes[1, 1].set_xticklabels(pivot_ap.columns, rotation=0)
    axes[1, 1].set_yticks(range(pivot_ap.shape[0]))
    axes[1, 1].set_yticklabels(pivot_ap.index)
    axes[1, 1].set_title("AP promedio por Modelo/Estrategia")
    for i in range(pivot_ap.shape[0]):
        for j in range(pivot_ap.shape[1]):
            val = pivot_ap.values[i, j]
            axes[1, 1].text(j, i, f"{val:.3f}" if not np.isnan(val) else "‚Äî",
                            ha="center", va="center", color="white" if not np.isnan(val) else "gray")

    # 6) Umbral √≥ptimo por label
    axes[1, 2].barh(top["label"], top["opt_threshold"])
    axes[1, 2].invert_yaxis()
    axes[1, 2].set_xlabel("Umbral √≥ptimo (por F1)")
    axes[1, 2].set_title("Distribuci√≥n de umbrales √≥ptimos")
    axes[1, 2].grid(axis="x", alpha=0.3)

    plt.tight_layout()

    # ---------- Logging a MLflow ----------
    # Si hay un run activo, logueo nested; si no, creo uno nuevo
    active = mlflow.active_run()
    nested = active is not None

    with mlflow.start_run(run_name=run_name, nested=nested):
        mlflow.set_tags({
            "summary_for": "balancing_comparison",
            "primary_metric": primary_metric,
            "has_winsor_col": str("winsor" in df.columns)
        })

        # Figura
        try:
            mlflow.log_figure(fig, f"{artifact_prefix}.png")
        except Exception:
            # fallback
            out_file = f"{artifact_prefix}.png"
            fig.savefig(out_file, dpi=300, bbox_inches="tight")
            mlflow.log_artifact(out_file)

        # CSVs
        df.to_csv(f"{artifact_prefix}_leaderboard.csv", index=False)
        agg.to_csv(f"{artifact_prefix}_by_strategy.csv", index=False)
        mlflow.log_artifact(f"{artifact_prefix}_leaderboard.csv")
        mlflow.log_artifact(f"{artifact_prefix}_by_strategy.csv")

        # M√©tricas agregadas por estrategia
        for _, r in agg.iterrows():
            s = r["imbalance_strategy"]
            mlflow.log_metric(f"mean_AP_{s}", float(r["mean_AP"]))
            mlflow.log_metric(f"mean_AUC_{s}", float(r["mean_AUC"]))
            mlflow.log_metric(f"mean_F1opt_{s}", float(r["mean_F1opt"]))
            mlflow.log_metric(f"mean_Popt_{s}", float(r["mean_Popt"]))
            mlflow.log_metric(f"mean_Ropt_{s}", float(r["mean_Ropt"]))
            mlflow.log_metric(f"n_{s}", int(r["n"]))

        # M√©tricas globales
        mlflow.log_metric("n_total_rows", int(len(df)))
        mlflow.log_metric("best_primary_value", best_primary_value)
        mlflow.log_metric("avg_AP_overall", float(df["avg_precision_test"].mean()))
        mlflow.log_metric("avg_AUC_overall", float(df["roc_auc_test"].mean()))
        mlflow.log_metric("avg_F1opt_overall", float(df["f1_at_opt_thr"].mean()))

        # Params del ganador
        mlflow.log_param("best_label", best_label)
        mlflow.log_param("best_model", best_model)
        mlflow.log_param("best_strategy", best_strategy)
        mlflow.log_param("best_by", primary_metric)

    plt.show()
    print("üìä Visualizaciones y resumen registrados en MLflow.")
    return {
        "best_row": best_row,
        "by_strategy": agg,
        "sorted": df_sorted
    }
````

````python
#5.10 ============================COMPARA ENTRE LOS GRUPO TREES y OTHERS (LR; SVC, KNN; NB)

def compare_by_group_balancing(df_leader: pd.DataFrame,
                               primary_metric: str = "avg_precision_test",
                               artifact_prefix: str = "by_group_summary",
                               run_name: str = "BY_GROUP_SUMMARY"):

    if "model_group" not in df_leader.columns:
        raise ValueError("df_leader no tiene columna 'model_group'. Asegurate de setear ese tag en el loop.")

    # Best por modelo (ya est√° ordenado en tu pipeline, pero por las dudas reordenamos)
    df = df_leader.sort_values(primary_metric, ascending=False).copy()

    # Mejor por grupo
    best_per_group = df.loc[df.groupby("model_group")[primary_metric].idxmax()].reset_index(drop=True)

    # Agregado por grupo x estrategia
    agg = (df.groupby(["model_group", "imbalance_strategy"])
             .agg(mean_AP=("avg_precision_test", "mean"),
                  mean_AUC=("roc_auc_test", "mean"),
                  mean_F1opt=("f1_at_opt_thr", "mean"),
                  n=("model_name", "count"))
             .reset_index())

    # Log sencillo a MLflow
    with mlflow.start_run(run_name=run_name, nested=mlflow.active_run() is not None):
        best_per_group.to_csv(f"{artifact_prefix}_best_per_group.csv", index=False)
        agg.to_csv(f"{artifact_prefix}_agg.csv", index=False)
        mlflow.log_artifact(f"{artifact_prefix}_best_per_group.csv")
        mlflow.log_artifact(f"{artifact_prefix}_agg.csv")

        # m√©trica comparativa directa
        for _, r in best_per_group.iterrows():
            g = r["model_group"]
            mlflow.log_metric(f"best_{primary_metric}_{g}", float(r[primary_metric]))
            mlflow.log_param(f"best_model_{g}", f"{r['model_name']}|{r['imbalance_strategy']}")

    return {"best_per_group": best_per_group, "agg": agg}
````

````python
## 5.11 ============================= AGREGADO DE CATEGORIA CLASS_WEIGTH (Para modelos que lo usan) =============================
#Para √°rboles/boosters, adem√°s de SMOTE/NearMiss, suele valer probar una pasada con class_weight='balanced' (RF/DT) o scale_pos_weight (XGB/LGBM/CatBoost). Si quer√©s, agregamos una cuarta estrategia "CLASS_WEIGHT" s√≥lo para esos modelos sin tocar los dem√°s (o lo dejamos para el pr√≥ximo barrido).
#Con esto compar√°s todos tus candidatos (otros vs √°rboles) en el mismo experimento, con res√∫menes globales y por grupo, sin duplicar runs.

def apply_class_weight(clf, pos_weight: float):
    """
    Intenta activar class weighting en el clasificador clonado:
    - Sklearn (LogReg, SVC, Tree, RF): class_weight='balanced'
    - XGBoost/LightGBM/CatBoost: scale_pos_weight = n_neg / n_pos
    Devuelve: (clf_modificado, applied: bool, details: dict)
    """
    applied = False
    details = {}
    try:
        params = clf.get_params(deep=False)
    except Exception:
        params = {}

    try:
        # 1) Modelos sklearn con class_weight
        if "class_weight" in params:
            clf.set_params(class_weight="balanced")
            applied = True
            details["class_weight"] = "balanced"

        # 2) Boosters con scale_pos_weight
        elif "scale_pos_weight" in params:
            clf.set_params(scale_pos_weight=pos_weight)
            applied = True
            details["scale_pos_weight"] = float(pos_weight)

        # 3) CatBoost (por si no aparece en get_params)
        elif clf.__class__.__name__ == "CatBoostClassifier":
            try:
                clf.set_params(scale_pos_weight=pos_weight)
                applied = True
                details["scale_pos_weight"] = float(pos_weight)
            except Exception:
                pass
    except Exception:
        pass

    return clf, applied, details
````

## Lectura de resultados (qu√© mirar)

AP_test (Average Precision) y ROC_AUC_test: desempe√±o global.

F1@opt, Prec@opt, Rec@opt, thr_opt: qu√© umbral te conviene si quer√©s balancear precisi√≥n/recall.

Matriz: FN y TP te dicen si levanta recall de la minor√≠a.

Regla pr√°ctica

Si A (weights) > B (NearMiss) en AP/recall ‚Üí quedate con A (m√°s datos, menos riesgo de overfitting).

Si B mejora mucho recall (y te importa detectar positivos) sin destruir precisi√≥n, puede valer la pena, sobre todo en LogReg/SVC/KNN.
