```python
# Ejercicio n°3 Binomial
# Un curso de Estadística de la Universidad de Buenos Aires posee 10 alumnos, 
# de los cuales 4 son mujeres y el resto hombres. 
# Se eligieron al azar 7 estudiantes del curso para tomarles un examen. 
# Hallar la probabilidad de que, entre los alumnos seleccionados, haya tres mujeres.

from scipy.stats import binom

N = 10  # Total de alumnos
M = 4   # Total de mujeres
K = 7   # Total de seleccionados
x = 3   # Mujeres seleccionadas

p = M / N  # Probabilidad de seleccionar una mujer
q = 1 - p  # Probabilidad de seleccionar un hombre

# Probabilidad de seleccionar 3 mujeres y 4 hombres
prob = binom.pmf(x, K, p)
print(f"La probabilidad de seleccionar {x} mujeres entre {K} alumnos es: {prob:.4f}")

```

```python
#graficar esta distribucion de probabilidad
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
import seaborn as sns
import pandas as pd

# Definimos los parámetros de la distribución binomial
n = 10
p = 0.4  # Probabilidad de éxito (mujeres)
# Definimos el rango de valores para x (número de éxitos)
x = np.arange(0, n+1)
# Calculamos la función de masa de probabilidad (PMF)
pmf = binom.pmf(x, n, p)
# Creamos un DataFrame para facilitar la visualización
df = pd.DataFrame({'x': x, 'PMF': pmf})
# Creamos el gráfico
plt.figure(figsize=(10, 6))
sns.barplot(x='x', y='PMF', data=df, color='blue')
plt.title('Distribución Binomial PMF (n=15, p=0.34)')
plt.xlabel('Número de mujeres seleccionadas')
plt.ylabel('Probabilidad')
plt.xticks(x)
plt.grid(axis='y')
plt.show()
```
