```python
# Import required library
import pandas as pd
import matplotlib.pyplot as plt

# Define discrete probability distribution
dxi = [0, 1, 2, 3, 4, 5]  # Random variable values
Pxi = [2/15, 1/3, 0, 1/5, 4/15, 1/15]  # Corresponding probabilities

# Verify probabilities sum to 1
prob_sum = sum(Pxi)
assert abs(prob_sum - 1.0) < 1e-10, "Probabilities must sum to 1"

# Create dictionary mapping values to probabilities
distribution = dict(zip(dxi, Pxi))


# Create DataFrame from distribution
df = pd.DataFrame({
    'Value (x)': dxi,
    'Probability P(x)': Pxi
})

# Display probability distribution table
print("\nProbability Distribution:")
print(df.to_string(index=False))

# Create bar plot of probability distribution
ax = df.plot(
    kind='bar',
    x='Value (x)', 
    y='Probability P(x)',
    title='Probability Mass Function',
    xlabel='x',
    ylabel='P(x)',
    figsize=(10,6)
)
# Save plot to file
plt.savefig('probability_distribution.png')
plt.show()

plt.close()

```

```python
import numpy as np

# Datos para CDF
x_cdf = [-1, 0, 1, 2, 3, 4, 5, 6]  # Extender el rango para visualización
F = [0, 2/15, 7/15, 7/15, 2/3, 14/15, 1, 1]

# Crear gráfica
plt.figure(figsize=(8, 5))
plt.step(x_cdf, F, where='post', color='orange', linewidth=2.5)

# Añadir puntos en los saltos
plt.scatter(x_cdf[1:-1], F[1:-1], color='orange', s=50, zorder=5)

# Etiquetas
plt.xticks(np.arange(-1, 7))
plt.yticks(np.arange(0, 1.2, 0.2))
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Función de Distribución Acumulada (CDF)")
plt.grid(True, linestyle='--', alpha=0.6)

# Guardar como imagen
plt.savefig("cdf_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```
