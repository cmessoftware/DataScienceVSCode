```python
from scipy.stats import hypergeom

# Parámetros:
# M = tamaño de la población total
# n = cantidad de éxitos en la población
# N = tamaño de la muestra
N = 20     # Total de elementos
r = 2      # Elementos exitosos
n = 3      # Tamaño de la muestra

# ¿Cuál es la probabilidad de obtener exactamente k=3 éxitos?
k = 2
prob = hypergeom.pmf(k, N, r, n)
print(f"P(X = 3): {prob:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, N+1)
probs = hypergeom.pmf(x, N, r, n)

plt.bar(x, probs)
plt.title(f"Hipergeométrica: N={N}, r={r}, n={n}")
plt.xlabel("Éxitos en la muestra")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```

```python
# Simular 1000 experimentos de extracción
samples = hypergeom.rvs(N, r, n, size=1000)

plt.hist(samples, bins=np.arange(0, n+2)-0.5, edgecolor='black')
plt.title("Simulación Hipergeométrica")
plt.xlabel("Éxitos en la muestra")
plt.ylabel("Frecuencia")
plt.show()
```
