# Three special exponential families

In these notes, we discuss two special exponential families generated by
stochastic processes. The first process is the Poisson process, while the second process is a Gaussian process, while the final one is the empirical process.

## Poisson process

A Poisson process is a point process on some measurable space $(X, {\cal X})$. We assume that $X$ has some topology and ${\cal X}={\cal B}(X)$, the Borel sigma algebra.

The process determined by the intensity a measure on $(X, {\cal X})$ yielding a triple $(X, {\cal X},\mu)$. 
For the time being, we'll assume that $\mu(X)<\infty$.

The Poisson process is a random element of
$$
{\cal M}(X, {\cal X}) = \left\{q: \text{$q$ is a measure on $(X, {\cal X})$} \right\}
$$
and can be written as
$$
N = \sum_{t \in {\cal P}} \delta_t
$$
where ${\cal P} \subset X$ is the set of points of the process.

We call its distribution $\Pp_0$, which is a distribution on ${\cal M}(X, {\cal X})$. Formally, we should specify the sigma algebra on ${\cal M}(X, {\cal X})$, we'll call it ${\cal F}$.

Simple functions 
$$
f =  \sum_{i=1}^k c_i 1_{A_i}
$$
can be integrated against $N$ by
$$
\begin{aligned}
N(f)
&= \sum_{i=1}^k c_i \cdot \# ({\cal P} \cap A_i) \\
 &= \sum_{i=1}^k c_i \cdot N(A_i) \\
    &= \langle f, N \rangle
\end{aligned}
$$
where the last line is meant to denote a pairing between functions on
$(X,{\cal X})$ and measures on $(X, {\cal X})$.

Poisson processes have the special property that if 
$A_1, \dots, A_k$  are pairwise disjoint, then 
$$
N(A_i) \overset{\text{indep}}{\sim} \text{Poisson}(\mu(A_i)).
$$


Therefore,
$$
\begin{aligned}
\Ee_0 \left(e^{\langle f, N\rangle} \right) &= \prod_{i=1}^k e^{\mu(A_i) (e^{c_i}-1)} \\
&= \prod_{i=1}^k e^{\mu(A_i) (e^{c_i}-1)} \\
&= \exp \left(\int_X \left[e^{f(x)} - 1\right] \mu(dx) \right)
\end{aligned}
$$



The above formula holds for simple functions whose sets are pairwise disjoint. Any simple function can be reexpressed in this form, so we can now compute the moment generating function of any simple function.
The integral can now be extended to more general functions.

### Construction of the Poisson process

The following generates a realization of a Poisson process with intensity $(X, {\cal X},\mu)$:
    
* Generate $Z \sim \text{Poisson}(\mu(X))$.

* Given $Z$, generate $X_i \overset{IID}{\sim} Q, 1 \leq i \leq Z$ where $Q$ is the distribution
$$
Q(A) = \frac{\mu(A)}{\mu(X)}.
$$

* Set $N=\sum_{i=1}^Z \delta_{X_i}$.

Here's an example on the plane, with intensity measure given by
$$
\mu(A) = 125.3 \cdot \int_A e^{-\|x\|^2_2/2} \; dx
$$

```python
rate = 125
Z = np.random.poisson(rate)
X = np.random.standard_normal((2,Z))
plt.scatter(X[0],X[1])
plt.show()
```

Let's integrate a function against $N$.

```python
def integrate(f, X):
    return np.sum([f(x,y) for x,y in X.T])

def f(x,y):
    return 3*x*y+2*np.exp(x)

integrate(f, X)
```

We will also want to be able to compute the expected value of the integral under
$\Pp_0$.

```python
def E_0(f, rate, ndraws=10**4):
    X0 = np.random.standard_normal((2,ndraws))
    return rate * integrate(f, X0) / ndraws
```

Let's check that the mean is roughly 0.

```python
def check_mean(f, rate, ntrial=20):
    results = []
    for _ in range(ntrial):
        Z = np.random.poisson(rate)
        X = np.random.standard_normal((2,Z))
        integral = integrate(f, X)
        mean_estimate = E_0(f, rate)
        results.append(integral - mean_estimate)
    return np.array(results)

results = check_mean(f, rate)
print np.mean(results), np.std(results)
```

We claim this procedure produces a Poisson process. To check, we should verify the defining
property of this Poisson process, namely that for pairwise disjoint sets, the
resulting counts are independent Poisson with the correct rates.

### The exponential family

We are now ready to define the exponential family generated by 
$\Pp_0$. Above, we computed a candidate for the CGF for simple functions
$$
\CGF(f) = \Ee_0\left[e^{\langle f, N \rangle} \right] = \int_X \left[e^{f(x)}-1 \right] \; \mu(dx).
$$

Our candidate exponential family is
$$
\frac{d\Pp_f}{d\Pp_0} = \langle f, N \rangle - \CGF(f)
$$
in which the process $N$ is itself the sufficient statistic.

We take our natural parameter space to be $C_b(X)$ the set of bounded continuous functions on $X$.

Having a family, we might now ask about the mean space. For $f=0$, the mean of the sufficient statistic is the intensity measure $\mu$. Therefore, we expect
the mean values to also be intensity measures. The usual approach is to compute the gradient of $\CGF$.

In this setting, we can compute directional derivatives of $\CGF$
$$
\begin{aligned}
\lim_{t \rightarrow 0} \frac{1}{t}\left[\CGF(f+t \cdot g) - \CGF(f)\right] &= 
\lim_{t \rightarrow 0} \frac{1}{t}\int_X \left[ e^{f(x) + t \cdot g(x)} - e^{f(x)} \right] \mu(dx) \\
&= \int_X g(x) e^{f(x)} \; \mu(dx).
\end{aligned}
$$

If $\CGF$ is differentiable, the derivative should be a (continuous) linear mapping that takes a direction (i.e. $g \in C_b(X)$) and returns the derivative in this direction.

We see the derivative is in fact in ${\cal M}(X, {\cal X})$ and is the measure
with density $e^{f(x)}$ with respect to $\mu$:
$$
\frac{d\nabla \CGF(f)  }{d\mu}(x) = e^{f(x)}.
$$

Having found the mean value space, we can now find the analogy of $\CGF^*$.
For $q=e^f \cdot \mu$ define
$$
\begin{aligned}
\CGF^*(q) &= \sup_{h \in L^{\infty}(X, {\cal X}, \mu)} \langle h, q \rangle - \CGF(h) \\
&= \sup_{h \in C_b(X)} \int_X \left[h(x) e^{f(x)} \right] \mu(dx) - \int_X \left[ e^{h(x)} - 1\right] \mu(dx) \\
&= \sup_{h \in C_b(X)} \int_X \left[h(x) e^{f(x)} - e^{h(x)} + 1 \right] \mu(dx).
\end{aligned}
$$

The maximizer, $h^*(q)$ should satisfy
$$
q - \nabla \CGF(h^*(q)) = 0
$$
by which we see that $h=f$ and
$$
\CGF^*(q) = \int_X \left[f(x) e^{f(x)} - e^{f(x)} + 1 \right] \mu(dx)
$$

Now, we might ask about doing maximum likelihood estimation. In finite dimensions, we would 
find the MLE by
$$
\hat{f} = \nabla \CGF^*(N)
$$
where $N$ is the sufficient statistic. Unfortunately, $N$ is generally not measure with a bounded continuous density with respect to $\mu$ (it usually will not have a density with respect to $\mu$). So, the MLE is not defined.

In other words, we need a *model*...

### Processes derived from a Poisson process

Given a Poisson process, one can define a new stochastic process. For some parameter space $T$, suppose we have
$F:T \rightarrow C_b(X)$. This creates a new process
$$
f(t) = \langle F(t), N \rangle - \int_X F(t)(x) \mu(dx) , \qquad t \in T.
$$

This is generally called a *scan statistic* or a *scan statistic process*. 

It is called a scan statistic because, for each $t$ fixed
$f(t)$ is a score test of a certain alternative hypothesis $H_{a,t}$ about the intensity against $H_0$ which is that the intensity is $\mu$.

Specifically, suppose 
$$
H_{a,t}: \text{intensity has density $e^{\alpha F(t)}$ with respect to $\mu$}.
$$
The score test is
$$
\frac{d}{d\alpha} \biggl[\ell_t(\alpha) - \ell_t(0)\biggr]_{\alpha=0}
$$
where $\ell_t$ is the likelihood for $\alpha$.

We see 
$$
\ell_t(\alpha) - \ell_t(0) = \left[ \langle \alpha \cdot F(t), N \rangle - \int_X \left[e^{\alpha \cdot F(t,x)} - 1\right] \; \mu(dx) \right]
$$
and
$$
\frac{d}{d\alpha} \biggl[\ell_t(\alpha) - \ell_t(0)\biggr] = \langle F(t), N \rangle - \int_X F(t,x) e^{\alpha \cdot F(t,x)} \; \mu(dx)
$$
from which our claim follows.

Here's an example with $T=[-1,1]$:

```python
def F(t):
    return (lambda x,y: -np.clip(t,-1,1)*np.cos((x-t)**3) + np.clip(t,-1,1)*np.sin(y-t)**2)

T = np.linspace(-1,1,501)
f = [integrate(F(t), X) - E_0(F(t), rate, ndraws=10000) for t in T]
plt.plot(T, f)
```

The wiggiliness in the plot above is noise from estimating $\Ee_0(F(t))$. A better procedure would be to use the same noise to estimate $\Ee_0(F(t))$, i.e.
$$
\begin{aligned}
\widehat{\Ee_0(F(\cdot))} &= \frac{1}{M}\sum_{i=1}^M \int_X F(\cdot)(x) N_i(dx) \\
&= \int_X F(\cdot)(x) \left(\sum_{i=1}^M N_i(dx) \right)
\end{aligned}
$$
where $N_i$ are IID realizations of the null Poisson process.

This is easy to do by just simulating once from a Poisson process with intensity
$M \cdot \mu$.

```python
rate = 125
ndraws = 100
Z0 = np.random.poisson(rate * ndraws)
X0 = np.random.standard_normal((2,Z0))
f = [integrate(F(t), X) - integrate(F(t), X0) / ndraws for t in T]
plt.plot(T, f)
```

The above process is an example of a *curved exponential family*.

### Allowing for infinite rate

Above, we have assumed $\mu(X) < \infty$. Poisson processes do not need this assumption in general. Generally, if we replace $C_b(X)$ with $C_c(X)$, the set of continuous functions with compact support, then what we wrote above needs little modification.

## Gaussian processes

The next class of examples is the class of Gaussian processes. A ($\real$-valued) Gaussian process on a parameter space $T$ is a map $F:T \times \Omega \rightarrow \real$ such that,
for each $(t_1, \dots, t_k)$ the random vector
$$
\left(f(t_1), \dots, f(t_k)\right) \sim N\left(\begin{pmatrix} \mu(t_1) \\ \vdots  \\ \mu(t_k) \end{pmatrix}, \Sigma(t_1, \dots, t_k)\right)
$$
where
$$
\Sigma(t_1, \dots, t_k)_{ij} = R(t_i,t_j)
$$
with $R: T \times T$ the covariance function of $f$ which satisfies
$$
\sum_{i,j=1}^k a_i a_j R(t_i,t_j) \geq 0
$$
for all $((a_i,t_i), 1 \leq  i \leq k)$.


The simplest example of a Gaussian process is a Gaussian random vector in $\real^p$ which is
a process on $\{1, \dots, p \}$.
The covariance function $R:\{1,\dots,p\} \times \{1, \dots, p\} \rightarrow \real$ is just the matrix $\Sigma$.
    
A Gaussian random vector with distribution $N(\mu,\Sigma)$ generates an exponential family with $\Pp_0=N(\mu,\Sigma)$ and
$$
\frac{d\Pp_{\eta}}{d\Pp_0}(x) = \exp \left(\eta^Tx - \mu^Tx - \frac{1}{2} x^T\Sigma x \right)
$$
and the family is
$$
\left\{N(\mu_0 + \Sigma\eta, \Sigma) :\eta \in \real^p \right\}.
$$

From this we see that, apart from the choice of $\mu_0$, the set of allowable means is determined by $\Sigma$. Without loss of generality, we'll assume $\mu_0=0$ knowing we can always add $\mu_0$ to the process later.

The general picture is similar: the set of allowable means in our family is determined by $R$ in that it determines what is called the *reproducing kernel Hilbert space* determined by $R$ which we try to describe now.

In the finite dimensional setting, we can write
$$
\eta = \sum_{i=1}^p a_i e_i
$$
which  leads to a mean
$$
\mu = \sum_{i=1}^p a_i R[i]
$$
where $R[i]$ is the $i$-th row (or column) of $\Sigma$.

For this $\eta$ we see that (recalling $\mu_0=0$)
$$
\CGF \left(\sum_{i=1}^p a_i e_i \right) = \CGF(\eta) = \frac{1}{2} \eta^T \Sigma \eta = \frac{1}{2} \Vv_0 \left(\sum_{i=1}^p a_i f_i \right)
$$
where $f \sim N(0,\Sigma)$ is our Gaussian process on $\{1, \dots, p\}$.

The exponential family can then be written as
$$
\frac{d\Pp_{\eta}}{d\Pp_0}(x) = \exp \left(\sum_{i=1}^p a_i f_i - \frac{1}{2} \Vv_0 \left(\sum_{i=1}^p a_i f_i \right) \right)
$$
and the mean map is 
$$
\nabla \CGF(\eta) = \nabla \CGF \left(\sum_{i=1}^p a_i e_i \right) = \sum_{i=1}^p a_i R[i].
$$

This can be generalized to the general picture. Given $(a_i,t_i)_{1 \leq i \leq k}$ set
$$
\eta = \sum_{i=1}^k a_i \delta_{t_i},
$$
define
$$
W(\eta) = W\left(\sum_{i=1}^k a_i \delta_{t_i} \right) = \sum_{i=1}^k a_i f(t_i)
$$
and 
$$
\CGF\left(\sum_{i=1}^k a_i \delta_{t_i}\right) = \frac{1}{2} \Vv_0 \left( \sum_{i=1}^k a_i f(t_i) \right)
$$

### RKHS

The above exercise hints at the definition of the  *reproducing kernel Hilbert space (RKHS)*. 

There are actually three spaces in question here: the natural parameter space, the mean value space and the subspace of $L^2(\Pp_0)$ that contains all the limits of $W$ above. They are all isomorphic to the RKHS.
    

Let's start with the natural parameter space. 
Formally, we can define the RKHS as the completion of 
$$
\left\{ \eta: \eta = \sum_{i=1}^k a_i \delta_{t_i} \right\}
$$
with bilinear form
$$
\begin{aligned}
\langle \eta_1, \eta_2 \rangle_{H_N} &\overset{\text{def}}{=} 
\int_{T \times T} R(t,s) \; \eta_1(ds) \; \eta_2(dt) \\
&= \Ee_0 \left(W(\eta_1) W(\eta_2) \right) \\
\end{aligned}
$$
Completion means we include the limit of Cauchy sequences to form a Hilbert space.  The $H_N$ above denotes $N$ for natural parameter space.

The last equality above is the first isomorphism, yielding an isomorphism between all possible limits $W(\eta_m)$ and $H_N$. 

Therefore, in the natural parameter space the CGF is
$$
\CGF(\eta) = \frac{1}{2} \int_{T \times T} R(t,s) \; \eta(ds) \; \eta(dt) = \frac{1}{2} \|\eta\|^2_{H_N}
$$

What about the mean value space? Well,  we know the mean values should be a function, and should also satisfy
$$
\begin{aligned}
\langle \zeta, \nabla \CGF(\eta) \rangle &= \lim_{t \rightarrow 0} \frac{1}{t} \left[ \CGF(\eta + t \cdot \zeta) - \CGF(\eta)\right] \\
&= \langle \zeta, \eta \rangle_{H_N} \\
&= \int_T \left[\int_T R(t,s) \; \eta(ds) \right] \; \zeta(dt).
\end{aligned}
$$
where the first pairing is between functions and measures with the first argument a measure.

We see, then that
$$
\mu(t) = \langle \delta_t, \nabla \CGF(\eta) \rangle = \langle \delta_t, \eta \rangle_{H_N} = \int_T R(t,s) \; \eta(ds).
$$
This is just a generalization of the finite dimensional picture we saw above.



As we mentioned above, there is an isomorphism between the natural parameter picture and the mean space picture. This isomorphism is determined the map
$$
\nabla \CGF(\eta) = \int_T R(t,s) \eta(ds).
$$

Hence, on the range of $M$, we can define
$$
\begin{aligned}
\langle \mu_1, \mu_2 \rangle_H &= \langle \nabla\CGF^{-1}(\mu_1), \nabla\CGF^{-1}(\mu_2) \rangle_{H_N}.
\end{aligned}
$$



Or,
$$
\langle \nabla \CGF(\eta_1), \nabla \CGF(\eta_2) \rangle_H = \langle \eta_1, \eta_2 \rangle_{H_N} = \int_{T \times T} R(t,s) \; \eta_1(ds) \; \eta_2(dt).
$$

This inner product on functions is the one we usually refer to as the RKHS determined by $R$.

The *reproducing* in RKHS comes from the fact that the functions $R_t:T \rightarrow \real$ defined as
$$
R_t(s) = R(t,s)
$$
*reproduce* the kernel in some sense.

How? Note that 
$$
R_t = \nabla \CGF(\delta_t).
$$
Therefore,
$$
\langle R_t, R_s \rangle_H = \langle \delta_t, \delta_s \rangle_{H_N} = R(t,s).
$$

This property is used extensively in the *kernel trick* which we discuss below and we will revisit when we talk about regularized exponential families. It is essentially a type of interpolation property.

### The exponential family

We are now ready to define our exponential family. Our reference measure is
$\Pp_0$. Above, in our construction, we used a particular $\Pp_0$ but there may be others. In any case, it is a probability space on which we have a Gaussian process defined.

The natural parameter space is $H_N$ and
$$
\frac{d\Pp_{\eta}}{d\Pp_0} = \exp\left(W(\eta) - \frac{1}{2} \|\eta\|^2_{H_N} \right).
$$

Above
$$
W(\eta) = \sum_{i \geq 1}\xi_i \left[\int_T\phi_i(t) \; \eta(dt) \right],
$$
in the $L^2(\Omega, {\cal F}, \Pp)$ sense. Formally, we can also take $\xi_i=W(\nabla \CGF^{-1}(\phi_i))$.

So, we can take our sufficient statistic to be 
$$
\sum_{i \geq 1} \xi_i \phi_i.
$$



The process $f$ is still defined under $\Pp_{\eta}$. As expected, it is
again a Gaussian process and its mean is
$$
\mu(\cdot) = \left(\nabla \CGF(\eta)\right)(\cdot) =\int_T R(\cdot,s ) \; \eta(ds).
$$

How do we know this?
Well, we can compute moment generating functions of 
$$
\sum_{j=1}^k b_i f_{t_i}
$$
for an arbitrary choice of $(b_i,t_i)_{1 \leq i \leq k}$.

The moment generating function is:
$$
\begin{aligned}
\Ee_{\eta} \left(e^{\alpha \cdot \sum_{j=1}^k b_i f_{t_i}} \right) &= \Ee_{\eta} \left(e^{W \left( \alpha \cdot \sum_{j=1}^k b_i \delta_{t_i} \right)} \right) \\
&= \Ee_0 \left(e^{W \left(\eta + \alpha \cdot \sum_{j=1}^k b_i \delta_{t_i} \right) - \CGF(\eta)} \right) \\
&= \exp \left(\CGF\left(\eta + \alpha \cdot \sum_{j=1}^k b_i \delta_{t_i} \right) - \CGF(\eta) \right) \\
&= \exp \left(  \left \langle \eta, \alpha \cdot \sum_{j=1}^k b_i \delta_{t_i} \right \rangle_{H_N} + \CGF\left(\alpha \cdot \sum_{j=1}^k b_i \delta_{t_i}\right) \right) \\
&= \exp \left( \alpha \sum_{j=1}^k b_i \mu(t_i) + \frac{\alpha^2}{2} \sum_{j,l=1}^k b_i b_l R(t_j,t_l) \right)
\end{aligned}
$$
where
$$
\mu(\cdot) = \left(\nabla \CGF(\eta)\right) = \int_T R(t, \cdot) \mu(dt).
$$

This is the moment generating function of 
$$
N \left(\sum_{j=1}^k b_i \mu(t_i), \sum_{j,l=1}^k b_i b_l R(t_j,t_l) \right).
$$

So, under $\Pp_{\eta}$ the process is Gaussian with mean $\mu$ and covariance function $R$.

The set of reachable means, then, is $H$, the image of $H_N$ under 
$\nabla \CGF$. This formula is an infinite dimensional version of 
the finite dimensional picture where the set of reachable means is the rowspace
of $\Sigma$.

For $\mu \in H$, we see
$$
\begin{aligned}
\CGF^*(\mu) &= \sup_{\eta \in H_N} \langle \eta, \mu \rangle - \frac{1}{2} \|\eta\|^2_{H_N} \\
&= \sup_{\eta \in H_N} \langle \nabla \CGF(\eta), \mu \rangle_H - \frac{1}{2} \|\eta\|^2_{H_N} \\
&= \frac{1}{2} \|\mu\|^2_H
\end{aligned}
$$
with maximizing argument $\eta^*(\mu)$ satisfying
$$
\nabla \CGF(\eta^*(\mu)) = \mu.
$$


### MLE map

The MLE map is then a map from $H$ to $H_N$. As $\nabla \CGF$ is a linear map, we see that
$$
\nabla \CGF^*(\mu) = \nabla \CGF^{-1}(\mu).
$$

That is, it is just the inverse of $\nabla \CGF$ on its range, $H$.

Therefore, if we can perform MLE estimation based on the sufficient statistic $\sum_{i \geq 1} \xi_i \phi_i$ then
$$
\hat{\eta} = \nabla \CGF^{-1} \left(\sum_{i \geq 1} \xi_i \phi_i \right).
$$

Unfortunately,  $\sum_{i \geq 1} \xi_i \phi_i \not \in H$. To see this,
note that
$$
\left \|\sum_{i \geq 1} \xi_i \phi_i  \right\|^2_H = \sum_{i \geq 1} \xi_i^2 = +\infty
$$
whenever $H$ is infinite dimensional, as $W(\phi_i)_{i \geq 1}$ is a sequence of
IID $N(0,1)$ random variables. So, again, we need a *model*.

### Adding a different mean

Of course, since our process is a function, we can add an arbitrary
function $\mu_0:T \rightarrow \real$ to $f$ yielding a process
$$
g(t) = f(t) + \mu_0(t).
$$

We might ask what is the relation between exponential family determined by $g$ and the exponential family determined by $f$. If $g \in H$, then the families
are the same, while if $g \not \in H$ the two exponential families are different, in the sense that it is not possible to tilt any point in the family determined by $f$ to a point in the family determined by $g$.

### Tilting the covariance

In finite dimensions, the normal family has both mean and covariance parameters. We have just looked at tilting the mean. It is possible to consider the covariance as well, and the allowable covariances are also related to $R$. We will not go into details here, for lack of space.

### Example: Brownian motion

Let's look at Brownian motion in more detail. Brownian motion (with mean 0 and starting at 0) on $[0,1]$ is a
Gaussian process  $f$ with
$$
\Ee_0(f_t f_s) = \text{min}(t,s).
$$

The key object to identify is $H$, the RKHS of Brownian motion. From $H$, we can form the family easily. For Brownian motion (starting at 0 with mean 0)
$$
H = \left\{g:[0,1] \rightarrow \real, g(0)=0, \int_{[0,1]} \dot{g}^2(t) \; dt < \infty \right\}.
$$

Let's see if we can see why this is true.

Note that the form of $R$ shows that
$$
R_t(s) = \int_{[0,s]} 1_{[0,t]}(u) \; du = \int_{[0,1]} 1_{[0,s]}(u) 1_{[0,t]}(u) \; du.
$$
Or,
$$
\langle R_t, R_s \rangle_H = \langle 1_{[0,t]}, 1_{[0,s]} \rangle_{L^2([0,1])}
$$
which extends to
$$
\left \langle R_t, \sum_{i=1}^k a_i R_{s_i}  \right \rangle_H = \langle 1_{[0,t]}, \sum_{i=1}^k a_i 1_{[0,s_i]} \rangle_{L^2([0,1])}.
$$

For $h$ of the form
$$
h(t) = \sum_{i=1}^k a_i R_{s_i}(t)
$$
note that (in $L^2([0,1])$)
$$
\dot{h}(t) = \sum_{i=1}^k a_i 1_{[0,s_i]}(t)
$$
so the above reads
$$
h(t) = \langle R_t, h \rangle = \int_{[0,1]} 1_{[0,t]}(u) \dot{h}(u) \; du.
$$

We see the completion will admit any function $h$ such that
$\dot{h} \in L^2([0,1])$. We also see that $h(0)=0$.

The set of all reachable means then, in our exponential family, is
the set $H$, i.e. all functions whose derivative is square integrable and begin at 0.

### What is W?

Finally, what is $W$ in this setting? It is a linear map from $L^2([0,1])$ that
yields a Gaussian random variable. If we use the usual $B_t$ instead of $f_t$ for Brownian motion we see that
$$
W(h) = \int_{[0,1]} \dot{h}_t dB_t.
$$

That is, it is the (Ito) stochastic integral of a deterministic integrand with respect to Brownian motion. As we saw above, this $W$ in fact exists for general Gaussian processes.

For Brownian motion, one can integrate random integrands with respect to 
Brownian motion, one might ask whether it is possible to do the same for general Gaussian processes. The answer is yes, though we will not go into details.

### The kernel trick

Up to now, this Gaussian process material seems like abstract nonsense in that it has not helped us with estimation or anything else. 

Suppose, instead of observing the whole process, we only observed the process at
sites $(s_1, \dots, s_n) \in T$. Under $\Pp_{\eta}$ we know that
$$
(Y_1, \dots, Y_n) = \left(f(s_1), \dots, f(s_n)\right) \sim N(\mu_{\eta}(s_i), \Sigma)
$$
where $\mu_{\eta}=\nabla \CGF(\eta)$ and
$$
\Sigma^S_{ij} = R(s_i,s_j).
$$

We could ask what can we estimate about $\eta$ based on 
$(f(s_1), \dots, f(s_n))$?

The negative log-likelihood is, assuming $\Sigma^S$ is invertible,
$$
\frac{1}{2} \sum_{i,j=1}^n (Y_i - \mu_{\eta}(s_i))\left(\Sigma^S\right)^{-1}_{ij} (Y_j - \mu_{\eta}(s_j))
$$

The score equation is
$$
\Ee_{\eta} \left(f(s_i)\right) = Y_i, \qquad 1 \leq i \leq n
$$
so any $\eta$ for which
$$
\nabla \CGF(\eta)(s_i) = Y_i
$$
is an MLE. That is, any $\mu=\nabla \CGF(\eta)$ that interpolates the data exactly at $(s_1, \dots, s_n)$ is an MLE.

The point masses $\delta_{s_i}, 1 \leq i \leq n$ determine
a subspace 
$$
\tilde{L} = \tilde{L}(s_1, \dots, s_n) = \text{span}(\delta_{s_i}, 1 \leq i\leq n)
$$
of $H_N$. We can decompose any $\eta \in H_N$ as
$$
\eta = P_{\tilde{L}}\eta + P_{\tilde{L}}^{\perp}\eta.
$$

Equivalently, in $H$, we can write any $\mu \in H$ as
$$
\mu = P_L\mu + P_L^{\perp}\mu
$$
where this $L$ above is the image $\tilde{L}$ under $\nabla \CGF$.

### Nonparametric regression with kernels

This observation allows us to use kernels for nonparametric regression.
Let's start with squares regression in which we observe $(Y_i,s_i)_{1 \leq i \leq n}$ with
$$
Y_i|S \overset{\text{indep}}{\sim} N(\mu(s_i), \sigma^2 I)
$$
and want to find a good function $\mu$ to model the mean.

Formally, we are then trying to solve
$$
\hat{\mu} = \argmin_{\mu \in \real^T} \frac{1}{2 \sigma^2} \sum_{i=1}^n (Y_i - \mu(s_i))^2.
$$

This is ill-posed, in that any interpolator achieves 0 sum of squares. We know that for any RKHS $H$ there is an interpolator. There is also the interpolator with smallest $H$ norm. 

Here's an example with Brownian motion on $[0,1]$.

```python
def R(s,t):
    return np.minimum(s,t)

Y = [3,4,-2]
S = [0.2,0.4,0.9]

Gram = np.array([[R(s,t) for s in S] for t in S])
alpha = np.dot(np.linalg.pinv(Gram), Y)
```

Our estimator can be written as 

```python
def muhat(t):
    global alpha, S
    value = 0
    for a, s in zip(alpha, S):
        value += a * R(s, t)
    return value
```

The matrix 
$$
\Sigma^S_{ij} = R(s_i,s_j)
$$
is referred to as the Gram matrix for the kernel at the observed $s_i$'s.



Let's plot our function.

```python
tval = np.linspace(0,1,201)
plt.plot(tval, muhat(tval), label=r'$\hat{\mu}$')
plt.scatter(S, Y, c='red', s=30, label='data')
plt.legend()
```

### Regularized kernel estimators

An alternative is to consider a *regularized* estimator. For any RKHS, there is a natural penalty on $\mu$ the RKHS norm itself (or its square).  Consider then,
$$
\hat{\mu}_{\lambda} = \argmin_{\mu \in H} \frac{1}{2} \sum_{i=1}^n (Y_i - \mu(s_i))^2 + \frac{\lambda}{2} \|\mu\|^2_H = \argmin_{\mu\in H} L_{\lambda}(\mu)
$$

Decomposing $\mu$ as $P_L\mu + P_L^{\perp}\mu$ as above shows that the objective is actually equal to
$$
L_{\lambda}(\mu) = \frac{1}{2} \sum_{i=1}^n (Y_i - (P_L\mu)(s_i))^2 + \frac{\lambda}{2} \left(\|P_L(\mu)\|^2_H + \|P_L^{\perp}\mu\|^2_H \right).
$$

This shows that 
$$
L \ni \hat{\mu}_{\lambda} = \sum_{i=1}^n \hat{\alpha}_i(\lambda) R_{s_i}.
$$



### *Exercise*

Prove the above statements.

The problem therefore reduces to a *finite dimensional problem*
$$
\hat{\alpha}_{\lambda} = \argmin_{\alpha \in \real^n} \frac{1}{2} \|Y - \Sigma^S\alpha\|^2_2 + \frac{\lambda}{2} \alpha^T \Sigma^S \alpha.
$$

Computing the derivative with respect to $\alpha$ and setting it equal to 0 yields
$$
\left((\Sigma^S)^2 + \lambda \Sigma^S\right) \hat{\alpha}_{\lambda} = \Sigma^S Y.
$$
Or,
$$
\hat{\alpha}_{\lambda} = (\Sigma_S + \lambda I)^{-1} Y.
$$

Let's try this out with our Brownian motion example.


```python
def regularized_estimator(Y, S, R, L):
    Y = np.asarray(Y)
    n = Y.shape[0]
    Gram = np.array([[R(s,t) for s in S] for t in S])
    alpha = np.dot(np.linalg.pinv(Gram + L * np.identity(n)), Y)
    
    def f(t, S=S, R=R, L=L, alpha=alpha):
        value = 0
        for a, s in zip(alpha, S):
            value += a * R(s, t)
        return value
    
    return f

```

```python
param = 0.1
muhat_reg = regularized_estimator(Y, S, R, param)
tval = np.linspace(0,1,201)
plt.plot(tval, muhat_reg(tval), label=r'$\hat{\mu}_{%0.1e}$' % param)

param = 100
muhat_reg = regularized_estimator(Y, S, R, param)
tval = np.linspace(0,1,201)
plt.plot(tval, muhat_reg(tval), label=r'$\hat{\mu}_{%0.1e}$' % param)

plt.scatter(S, Y, c='red', s=30, label='data')
plt.legend()
```

As we expect, for large values of $\lambda$, the function is shrunk heavily towards 0.

```python
Y = np.array([np.random.binomial(1, np.exp(m)/(1+np.exp(m))) for m in eta(tval)])
```

We can reuse some of our previous code to fit this model. The first term in the loss
is the deviance of a logistic regression model with design matrix $\Sigma^S$. The second term is just a quadratic part.

In practice, there are generally much faster ways to solve these problems -- this code is for demonstration and is pretty inefficient.

```python
%run ../scripts/glm

class GLM_kernel(GLM):
    
    def __init__(self, Y, S, family, R, L):
        self.Sigma_S = np.array([[R(s,t) for s in S] for t in S])
        n = Y.shape[0]
        GLM.__init__(self, Y, np.hstack([np.ones((n,1)), self.Sigma_S]), family)
        self.R = R
        self.L = L
        self.S = S
        
    def value(self, beta):
        v1 = GLM.value(self, beta)
        alpha = beta[1:]
        v2 = self.L * (alpha * np.dot(self.Sigma_S, alpha)).sum() / 2.
        return v1 + v2
    
    def grad(self, beta):
        v1 = GLM.grad(self, beta)
        v1[1:] += self.L * np.dot(self.Sigma_S, beta[1:])
        return v1
    
    def hess(self, beta):
        v1 = GLM.hess(self, beta)
        v1[1:,1:] += self.L * self.Sigma_S
        return v1
    
    def fitted_function(self, beta):
        def f(t, R=self.R, alpha=beta.copy(), S=self.S.copy()):
            value = 0
            intercept = alpha[0]
            for a, s in zip(alpha[1:], S):
                value += a * R(t,s)
            return value + intercept
        return f
 
```

Let's try playing with the regularization parameter a little and see how the fit changes.

```python
for L in [0.001,0.1,0.5]: 
    model = GLM_kernel(Y, tval, binomial.bernoulli, R, L)
    alpha = fit(model, niter=40)

    f = model.fitted_function(alpha)
    plt.plot(tval, f(tval), label=r'$\hat{\eta}_{%0.1e}$'%L)
plt.plot(tval, eta(tval), label=r'$\eta$')
plt.legend(loc='upper left')
```

### Generalized additive models

Now that we have introduced a little regularization, we can talk about

$$

This is an example of a generalized additive model.

## The empirical process

Without going through all the details explicitly, a third special exponential family we might consider would be one generated by an *empirical process*. This is arguably a
more important process than either of the previous two examples...

This is similar to the Poisson process family, only instead of having a Poisson process for the points, we have a *binomial process*. That is, we assume we have a sample of $n$ IID points from some distribution $(X, {\cal X}, \Pp)$.

The analogy of the Poisson distribution for disjoint sets in the Poisson process is, for a partition $A_1, \dots, A_k$ of $X$ we have
$$
