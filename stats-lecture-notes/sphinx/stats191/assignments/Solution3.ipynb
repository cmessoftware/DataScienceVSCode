# Question 1

Power is an important quantity in many applications of statistics. This question investigates the power of a test in simple linear regression. In a simple linear regression setting, suppose the true slope of the regression line is $\beta_1$ and the true intercept is $\beta_0$.
If we assume  $\sigma$ is known, then we can test $H_0: \beta_0 + 66 \beta_1 =66$ using
$$
Z = \frac{\hat{\beta}_0 + 66 \hat{\beta}_1 - 66}{\sigma_{\hat{\beta}_0 + 66 \hat{\beta}_1}}
$$
where
$\sigma_{\hat{\beta}_0 + 66 \hat{\beta}_1}$ is the standard deviation of $\hat{\beta}_0 + 66 \hat{\beta}_1$.

The power of this test is a function of the true value $(\beta_0 + 66 \hat{\beta}_1)$ as well as
the accuracy of our estimate $\sigma_{\hat{\beta}_0 + 66 \hat{\beta}_1}$ and is defined as
$$
P(\text{$H_0$ is rejected}).
$$
 
As we change the true $\beta_0 + 66 \beta_1$, the probability we reject $H_0$ changes: if the true value of $\beta_0 + 66 \beta_1$ is much larger than 66 relative to $\sigma_{\hat{\beta}_0 + 66 \hat{\beta}_1}$, then 
we are very likely to reject $H_0$.

- **What rule would you use to determine whether or not you reject $H_0$ at level $\alpha=0.1$.**

We would reject $H_0$ if $|Z| > z_{0.95}=1.64$.

- **What is the distribution of our test statistic $Z$?
Show that the distribution depends on the values $\beta_0 + 66 \beta_1, \sigma_{\hat{\beta}_0 + 66 \hat{\beta}_1}$ and the 66 in our null $H_0$ via the quantity $(\beta_0 + 66 \beta_1  - 66) / \sigma_{\hat{\beta}_0 + 66 \hat{\beta}_1}$.
We call this quantity the non-centrality parameter. **

If we define $\delta = (\beta_0 + 66 \beta_1  - 66) / \sigma_{\hat{\beta}_0 + 66 \hat{\beta}_1}$, we can write
$$
\begin{aligned}
Z = Z-\delta + \delta
\end{aligned}
$$
where $Z-\delta \sim N(0,1)$. Therefore, $Z \sim N(\delta, 1).$ To see that $Z-\delta \sim N(0,1)$ note that
$$
Z - \delta = \frac{\hat{\beta}_0 + 66 \hat{\beta}_1 - \beta_0 - 66 \beta_1}{ \sigma_{\hat{\beta}_0 + 66 \hat{\beta}_1}}
$$
and, as we are in the simple linear regression model, we have the canonical form of a $Z$ statistic (which is like $T$ but we do not have to estimate the variance). These statistics have the form
$$
\frac{\hat{\theta} - \theta}{\sqrt(Var(\hat{\theta}))} \sim N(0,1).
$$

- **Plot the power of your test as your function of the non-centrality parameter.**

The power of our test is
$$
\begin{aligned}
P(|Z| > 1.64) &= P(|N(\delta,1)| > 1.64) \\
&= P(N(\delta,1) > 1.64) + P(N(\delta,1) < -1.64) \\
&= P(N(0,1) > 1.64-\delta) + P(N(0,1) < -1.64-\delta) \\
&= (1 - {\tt pnorm}(1.64-\delta)) + {\tt pnorm}(-1.64-\delta)
\end{aligned}
$$

```python
%load_ext rmagic
```

```python
%%R
delta = seq(-4,4,length=101)
power = function(delta) {(1 - pnorm(1.64-delta)) + pnorm(-1.64-delta)}
plot(delta, power(delta), type='l', lwd=2)
abline(h=0.7, lwd=2, lty=2, col='red')
abline(v=2.1)
```

- **Roughly how large does the non-centrality parameter have to be in order to achieve
power of 70%?**

From our plot, it seems that if $|\delta| \approx 2.1$ the power is 70%. A more careful calculation shows
that 2.165 is a more accurate answer.


```python
%%R
power(2.165)
```

The dataset `state.x77` in R contains the following statistics (among others) related to the 50 states
of the United States of America:

* `Population`: population estimate (1975)

* `Income`: per capita income (1974)

* `Illiteracy`: illiteracy (1970, percent of population)

* `HS.Grad`: percent high school graduates (1970)

```python
%%R
state.data = data.frame(state.x77)
```

We are interested in the relation between Income and other 3 variables.

- ** Produce a 4 by 4 scatter plot of the variables above.**

```python
%%R
pairs(state.data[,c('Population', 'Income', 'Illiteracy', 'HS.Grad')])
```

- **Fit a multiple linear regression model to the data with Income as the
dependent variable, and Population, Illiteracy, HS.Grad as the independent
variables. Comment on the significance of the variables in the model using the
result of summary.**

```python
%%R
income.lm = lm(Income ~ Population + Illiteracy + HS.Grad, data=state.data)
summary(income.lm)
```

It seems that `HS.Grad` and `Population` have fairly strong effects allowing for all other variables. 
It is possible that any effect of `Illiteracy` is subsumed by the `HS.Grad` variable as these are somewhat correlated.

- Produce standard diagnostic plots of the multiple regression fit in part 2.

```python
%%R
par(mfrow=c(2,2))
plot(income.lm)
```

- Plot dffits of the observations and find observations which have high influence,
using critical value 0.5.

```python
%%R
DFFITS = dffits(income.lm)
high_inf = DFFITS > 0.5
plot(DFFITS)
points(which(high_inf), DFFITS[high_inf], pch=23, bg='red')
state.data[high_inf,]
```

- **Plot Cook's distance of the observations and find observations which have high
influence, using critical value 0.1. Compare with the result of part 4.**

```python
%%R
COOKS = cooks.distance(income.lm)
high_inf = COOKS > 0.1
plot(COOKS)
points(which(high_inf), COOKS[high_inf], pch=23, bg='red')
state.data[high_inf,]
```

Alaska remains influential but North Dakoto is no longer considered influential. 

- **Find states with outlying predictors by looking at the leverage values. Use
critical value 0.3.**

```python
%%R
LEV = hatvalues(income.lm)
high_inf = LEV > 0.3
plot(LEV)
points(which(high_inf), LEV[high_inf], pch=23, bg='red')
state.data[high_inf,]
```

- **Find outliers, if any, in the response. Remove them from the data and refit a
multiple linear regression model and compare the result with the previous fit.**

```python
%%R
library(car)
outlierTest(income.lm)
subset.noAK = rep(TRUE, nrow(state.data))
subset.noAK[2] = FALSE
income.lm.noAK = lm(Income ~ Population + Illiteracy + HS.Grad, data = state.data, subset=subset.noAK)
summary(income.lm.noAK)
```

Based on the summary table, our conclusions are similar, though `Illiteracy` now seems to have a weak effect.

- **As a summary, find all the influential states using `influence.measures` function.**

```python
%%R
influence.measures(income.lm)
```

# Question 4

The	dataset	`iris`	in `R`	gives the measurements in centimeters of the
variables sepal length and width and petal length and width, respectively, for 
50 flowers from each of 3 species of iris. 

```python
%%R
data(iris)
```

- **Fit a multiple linear regression model to the data with sepal length as 
the dependent variable and sepal width, petal length and petal width 
as the independent variables. **

```python
%%R
iris.lm = lm(Sepal.Length ~ Sepal.Width + Petal.Length + Petal.Width, data=iris)
summary(iris.lm)
```

- ** Test the reduced model of $H_0: \beta_{\tt sepal width}=\beta_{\tt petal length} = 0$
with an
F-test at level $\alpha=0.05$. **

```python
%%R
reduced.lm = lm(Sepal.Length ~ Petal.Width, data=iris)
anova(reduced.lm, iris.lm)
```

The $p$-value is tiny. We strongly reject $H_0$ at level $\alpha=0.05$.

- ** Test $H_0: \beta_{\tt sepal width} = \beta_{\tt petal length}$ at level $\alpha=0.05$**

There are a few ways to test this model. One is via an $F$-test, the other is a $t$-test. The $F$ test is most straightforward. The reduced model is
$$
Sepal.Length_i = \beta_0  + \beta_1 Petal.Width_i + \beta_2 (Sepal.Width_i + Petal.Length_i) + \epsilon_i.
$$


```python
%%R
iris$Z = iris$Sepal.Width + iris$Petal.Length
reduced.lm = lm(Sepal.Length ~ Z + Petal.Width, data=iris)
anova(reduced.lm, iris.lm)
```

We cannot reject the null hypothesis at level $\alpha=0.05$.

- **Test $H_0: \beta_{\tt sepal width} < \beta_{\tt petal length}$ at level $\alpha=0.05$.**

The one-sided test will have $p$-value that depends on the sign of $\hat{\beta}_{\tt sepal width} - \hat{\beta}_{\tt petal length}$. We will reject for large positive values of this quantity. In this case, our estimate is negative, so we will definitely not reject $H_0$.

```python
%%R
coef(iris.lm)['Sepal.Width'] - coef(iris.lm)['Petal.Length']
```
