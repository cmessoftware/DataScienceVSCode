- **Fit a linear regression model to the observed counts as a linear function of `T` and `P`.**

```python
%%R
asthma = read.table('http://stats191.stanford.edu/data/asthma.table', header=TRUE)
asthma.lm = lm(Y ~ T + P, data=asthma)
summary(asthma.lm)
```

- **The outcomes are counts, for which a common model is the so-called Poisson model which says that

```python
%%R
W = 1/fitted(asthma.lm)
asthma.lm.W = lm(Y ~ T+P, weights=W, data=asthma)
summary(asthma.lm.W)
```

- ** A simplified version of the random walk model of stock prices states that the best prediction of the stock index at `Time=t` is the value of the index at `Time=t − 1`. Show that this corresponds to a simple linear regression model for 2. with an intercept of 0 and a slope of 1.**

Let $Y_t = DJIA_t, X_t= DJIA_{t-1}$.
Then, the random walk model says that
$$
E(Y_t|X_t)= X_t = \beta_0 + \beta_1 X_t
$$
when $\beta_0=0, \beta_1=1$.

- ** Carry out the the appropriate tests of significance at level `α = 0.05` for 4. Test each coefficient separately ($t$-tests) , then test both simultaneously (i.e. an F test). **

We use the unlogged model here. The $t$ statistic for $H_0:\beta_0=0$ can be read
off the summary table above, though we recompute it using the usual $T$ statistic formula below.

The $t$ statistic for $H_0:\beta_1=1$ cannot be read directly off the summary
table, but it can be computed directly off the summary table by subtracting 1 from the estimate and dividing by the standard error of $\hat{\beta}_1$.

```python
%%R
T1 = coef(dowlag.lm)['(Intercept)'] / sqrt(vcov(dowlag.lm)['(Intercept)', '(Intercept)'])
P1 = 2 * (1 - pt(abs(T1), 259))
print(data.frame(T1,P1))

T2 = (coef(dowlag.lm)['X'] - 1) / sqrt(vcov(dowlag.lm)['X', 'X'])
P2 = 2 * (1 - pt(abs(T2), 259))
print(data.frame(T2,P2))
```

The $F$ test can be done by forcing the coefficients in the regression model to be $\beta_0=0, \beta_1=1$. We will do this via an `offset` term.

```python
%%R
dowreduced.lm = lm(Y ~ -1, offset=X)
anova(dowlag.lm, dowreduced.lm)
```

In any case, at level $\alpha=0.05$ we fail to reject any of the above hypotheses. So, the random walk model seems to be a reasonable model as we
cannot reject based on the data.

## Question 3

In this question we will look at inference ($t$-tests and confidence
intervals) after model selection. We will use the `prostate` data from `ElemStatLearn`.

```python
%%R
library(ElemStatLearn)
data(prostate)
```

We will perform 1 step of forward stepwise model selection 
when the outcome is generated completely independently from
the design matrix. Following this, we will look at the corresponding $t$-statistics and confidence intervals.

I've written 

```python
%%R
select_first = function(dataset) {
    dataset$Y = rnorm(nrow(dataset))
    first_model = step(lm(Y~1,data=dataset), list(upper=~lcavol + 
                                                  lweight + age + 
                                                  lbph + svi + 
                                                  lcp + pgg45),
                      trace=FALSE, k=0, steps=1)
    return(list(interval=confint(first_model)[2,], pvalue=summary(first_model)$coef[2,4]))
}
select_first(prostate)
```

** I have written a function above that you can use for this exercise.
By reading `help(step)` if necessary, describe what the function `select_first` does.**

This function finds the best single variable model. By setting `k=0` the "AIC" score it computes is essentially just the error sum of squares of the model. So,
at the first step it will pick the best model with one variable in it. Since we set `steps=1`, it will stop after taking one step.

After this, it returns the confidence interval for the chosen variable and
the corresponding $p$-value that tests when this is 0 or not.

** Collect a sample of 1000 realizations of `pvalue` by running `select_first` 1000 times and storing only the `$pvalue` entry. Plot a histogram of these 1000 $p$-values. What do you expect it to look like? What does it look like? Why? **

Naively, if the $p$-values were to be trusted, they should have a roughly uniform distribution. However, we are choosing the variable that has the smallest $p$-value. Therefore, we expect these $p$-values will, on average, be smaller than a Unif(0,1) random variable. So, the histogram should not be flat and should put more weight near 0.

```python
%%R
P = c()
for (i in 1:1000) {
    P = c(P, select_first(prostate)$pvalue)
}
hist(P)
```

** What are the *true* regression coefficients in the above simulation? What value do you expect each of the above `interval` realizations to cover? Collect 1000 realizations of `interval`.  How many do you expect to cover the true value? How many actually do cover the true value? **

The true regression coefficients are 0 above. So, if the intervals were valid
we would expect them to cover 0. So, out of 1000 intervals, at level $\alpha=0.05$, we expect roughly 950 to cover 0.

```python
%%R
count = 0
for (i in 1:1000) {
    I = select_first(prostate)$interval
    count = count + (I[1] < 0) * (I[2] > 0)
}
count
```

We see that the intervals fail to cover at a 95% level. The coverage is closer to 75%. So, these intervals cannot be trusted, at least when the true coefficient is 0.

** Repeat 2. and 3. but modify `select_first` to add a strong effect for the variable `lcavol`.**

Adding a strong effect is one in which the $t$ statistic for `lcavol` at the first step will be quite large. A true coefficient of 2 yields (for this realization of $Y$) a $t$ statistic of size 20 or so. I would consider that a pretty strong effect.

```python
%%R
Y = rnorm(nrow(prostate)) + 2*prostate$lcavol
summary(lm(Y ~ lcavol, data=prostate))
```

To modify the function, we just add `lcavol` when generating `Y`. This function
will fail when given a `dataset` that does not have this variable, but we only
need this function for this question.

```python
%%R
select_first_strong = function(dataset) {
    dataset$Y = rnorm(nrow(dataset)) + 2*dataset$lcavol
    first_model = step(lm(Y~1,data=dataset), list(upper=~lcavol + 
                                                  lweight + age + 
                                                  lbph + svi + 
                                                  lcp + pgg45),
                      trace=FALSE, k=0, steps=1)
    return(list(interval=confint(first_model)[2,], pvalue=summary(first_model)$coef[2,4]))
}
```

Now, we expect to almost always choose `lcavol` and, since the effect is relatively strong. So, the $p$-values returned by `select_first_strong` should all be very small (since the true coefficient is non-zero and the $t$ statistic is quite large). 

If we always choose `lcavol` with `select_first_strong`, then we are not doing any selection, and we would expect the intervals to have closer to 95% coverage.
In this case, the intervals should cover the true coefficient which we have 
fixed to be 2.

```python
%%R
count = 0
P = c()
for (i in 1:1000) {
    S = select_first_strong(prostate)
    I = S$interval
    count = count + (I[1] < 2) * (I[2] > 2)
    P = c(P, S$pvalue)
}
hist(P)
count
```

As expected, all the $p$-values are tiny, and the count is roughly 950.
